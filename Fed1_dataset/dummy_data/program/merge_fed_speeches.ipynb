```python
import pandas as pd
```

```python
df = pd.read_csv('./../data/speeches.csv')
df
```

```python
# null값이 있어 아래와 같이 처리
df['text'] = df['text'].astype(str).fillna('')

# '날짜' 열을 날짜 시간으로 변환하여 형식 및 처리 오류를 자동으로 추론
df['date'] = pd.to_datetime(df['date'], errors='coerce')

df = df.dropna(subset=['date'])
#text	index	speaker	date	title로 구성된 데이터프레임을 date,title,speaker별 text로 된 데이터 프레임으로 재구성
df_group = df.groupby(['date', 'title', 'speaker'])['text'].apply(lambda x: ' '.join(x)).reset_index()

df_group = df_group.sort_values(by=['date', 'title', 'speaker'])

df_group = df_group[['date', 'title', 'speaker', 'text']]
df_group
```

```python
df['text'] = df['text'].astype(str).fillna('')

# Convert 'date' column to datetime, coercing errors to NaT (Not a Time)
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Drop rows where 'date' is NaT
df = df.dropna(subset=['date'])

# Group by 'date', 'title', and 'speaker', and concatenate 'text' values
df_group = df.groupby(['date', 'title', 'speaker'])['text'].apply(lambda x: ' '.join(x)).reset_index()

# Sort by 'date'
df_group = df_group.sort_values(by=['date'])

# Reorder columns if needed
df_group = df_group[['date', 'title', 'speaker', 'text']]
```

```python
df_group.to_csv(f'./../data/fed_speech_merge.csv')
```
