```python
import pandas as pd
```

```python
file_path = './../Fed1_dataset/beige_book_2008_2024.xlsx'

# FED_ir의 의미 : FED interest rate
df_beigebook = pd.read_excel(file_path, engine='openpyxl')

df_beigebook.head()
```

```python
df_beigebook.dtypes
```

```python
# 연월로 된 데이트 컬럼 추가
df_beigebook['DATE_YM'] = (df_beigebook['date']).dt.to_period('M').astype(str) #  - pd.DateOffset(months=1)
df_beigebook.head()
```

```python
df_beigebook['DATE_YM'].value_counts().max() # 1개 이상인지 확인
```

```python
# 시계열 데이터와의 통합을 위해 Date=>DATE로 변경하고 object type으로 변경
df_beigebook = df_beigebook.rename(columns={'date': 'DATE'})
df_beigebook['DATE'] = df_beigebook['DATE'].astype(str)
df_beigebook.head()
```

```python
df_beigebook.dtypes
```

```python
df_beigebook
```

```python
df_irp = pd.read_csv('./../Fed1_dataset/dummy_data/data/Interest rate prediction.csv')
df_irp.head()
```

```python
# 'DATE' 열을 날짜 형식으로 변환
df_irp['DATE'] = pd.to_datetime(df_irp['DATE'])

# 'DATE_YM' 열을 생성하고 연월 형식으로 변환
#df_irp['DATE_YM'] = df_irp['DATE'].dt.to_period('M').astype(str)

# 'DATE_YM' 열을 전월로 계산하여 업데이트
df_irp['BB_text'] = ''
df_irp['BB_neg'] = ''
df_irp['BB_neu'] = ''
df_irp['BB_pos'] = ''
df_irp['BB_sentiment'] = ''
df_irp['DATE_YM'] = (df_irp['DATE']).dt.to_period('M').astype(str) #  - pd.DateOffset(months=1)
df_irp.head()
```

```python
df_irp.dtypes
```

```python
df_irp['DATE'] = df_irp['DATE'].astype(str) # data type object로 원복
```

```python
df_irp.dtypes
```

```python
df_irp
```

```python
for i in range(len(df_irp)) :
    for j in range(len(df_beigebook)) :
        if (df_irp.iloc[i]['DATE'] == df_beigebook.iloc[j]['DATE']) :
            df_irp.loc[i, 'BB_text']       = df_beigebook.loc[j, 'beige_book']
            df_irp.loc[i, 'BB_neg']        = df_beigebook.loc[j, 'neg']
            df_irp.loc[i, 'BB_neu']        = df_beigebook.loc[j, 'neu']
            df_irp.loc[i, 'BB_pos']        = df_beigebook.loc[j, 'pos']
            df_irp.loc[i, 'BB_sentiment']  = df_beigebook.loc[j, 'sentiment']
```

```python
df_irp
```

```python
df_irp.to_csv("./../Fed1_dataset/Interest_rate_prediction.csv",index=False)
```
